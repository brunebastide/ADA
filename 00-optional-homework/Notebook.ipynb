```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%load_ext autoreload
%autoreload 2
sns.set_palette('Blues')
sns.set_context("notebook")
```

```python
# import the dataset
dataset = pd.read_csv('interactive_data.csv')
dataset
```

```python
# death reasons
pd.unique(dataset['Intent'])
```

```python
suicide_only = dataset[dataset['Intent'] == 'Suicide']
homicide_only = dataset[dataset['Intent'] == 'Homicide']
accident_only = dataset[dataset['Intent'] == 'Accident']
explained_only = pd.concat([suicide_only, homicide_only, accident_only])
```

```python
# Slide 1
total_death = explained_only['Deaths'].sum()
print('Total number of "explained" deaths:', total_death)
```

```python
total_suicide = dataset['Deaths'][dataset['Intent'] == 'Suicide'].sum()
print('Total number of suicides:', total_suicide)
proportion = total_suicide / total_death
print('Proportion:', proportion * 100, '%')
```
