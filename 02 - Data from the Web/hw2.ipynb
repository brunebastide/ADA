```python
len(top_uni_json['data'])
```

```python
def json_to_dict(js):
    return {
        'name': js['title'],
        'country': js['country'],
        'rank': js['rank_display'],
        'url': js['url'],
        'region': js['region']
    }
```

```python
top_uni = list(map(json_to_dict, top_uni_json['data']))
top_uni[0]
```

We are still missing the fields number of faculty members (international and total) and number of students (international and total). We will download the details page of each university to extract these figures. `log_progress` is a utility function imported from the current directory to show the progress of the download.

```python
prefix = 'https://www.topuniversities.com'

# fields and associated HTML class to extract
fields_html = {
    'total_faculty': 'total faculty',
    'international_faculty': 'inter faculty',
    'total_students': 'students-main',
    'international_students': 'int-students-main'
}

for uni in log_progress(top_uni):
    url = prefix + uni['url']
    response = requests.get(url)
    page = BeautifulSoup(response.text, 'html.parser')
    for field, html_class in fields_html.items():
        try:
            value = locale.atoi(page.find(class_=html_class).find(class_='number').string.strip())
            uni[field] = value
        except Exception:
            uni[field] = np.nan
```

```python
top_uni_df = pd.DataFrame(top_uni)
top_uni_df['rank'] = np.arange(1, len(top_uni_df) + 1)
top_uni_df = top_uni_df.set_index(['name'])
top_uni_df.head()
```

# 3. Merge two rankings

In order to obtain a "perfect" merge, we will need to implement a matching algorithm in order to associate each university with itself eventhough it does not have the same spelling in each of the two rankings we are studying. For example "Ecole Polytechnique Fédérale de Lausanne" and "Ecole Polytechnique Fédérale de Lausanne (EPFL)" are note exactly the same-same. Here is how we will proceed: 

- we preprocess the universities names with `preprocess()` to lower all the letters, remove special characters, match special letters of different alphabets to ASCII,

- we replace most common words ('University'/'Universidad'/'Institute'/'Technology') by unique special characters (tokens), which lowers the resemblance ratio for some pair of names which do not belong to the same universities, e.g. 'University Utah' and 'University Ohio' become '# Utah' and '# Ohio'. We did not delete these words but replaced them by tokens to avoid confusion between different institutions in same locations, e.g. `[('tokyo', 100, 'University of Tokyo'), ('tokyo', 100, 'Tokyo Institute of Technology'), ('tokyo', 100, 'Tokyo University of Science')]`. 

- we then compute the resemblance ratio with difflib (see note below) between all universities names by country.

- finally we threshold the ratios to keep only the meaningful matches. Precisely, we cut at 88% match and check manually ratios between 70% and 88%.


The library `difflib` as stated earlier was very useful to compute the Levenshtein distance between some pair of names and thus computing their ratio of resemblance. We then needed to apply a threshold to accept matches having high ressemblance.

```python
from fuzzywuzzy import process
from fuzzywuzzy import fuzz
import re
from transliterate import translit
from transliterate.exceptions import LanguageDetectionError

top_uni_names = top_uni_df.index.to_series()
uni_200_names = uni_200_df.index.to_series()

def translit_(words):
    """match special characters of different alphabet back to ASCII"""
    try:
        return translit(words, reversed=True)
    except LanguageDetectionError:
        return words

def preprocess(serie):
    """uniformise the university names"""
    serie = serie.str.lower()
    pattern = re.compile('[\W _]+')
    # remove non letter/numbers
    serie = serie.apply(lambda u: pattern.sub(' ', u).strip())
    # replace special alphabet characters by simpler
    serie = serie.apply(translit_)
    return serie

def replace_words(serie, word_tokens):
    """replace words in serie by values in dict word_tokens if any"""
    return serie.apply(lambda u: ' '.join(word_tokens.get(w, w) for w in u.split(' ')))

top_uni_simpler = preprocess(top_uni_names)
uni_200_simpler = preprocess(uni_200_names)
```

```python
# find the most used words and tokenize them
words = pd.Series(' '.join(top_uni_simpler.append(uni_200_simpler)).split(' '))
n_stop_words = 15
tokens = '!@#$%^&*()_+{}|-=[]\:;<>?,./'
assert len(tokens) >= n_stop_words
stop_words = words.value_counts()[:n_stop_words].index.values
tokenized_stop = dict(zip(stop_words, tokens))
print('Tokenized stop words:', tokenized_stop)

top_uni_tokens = replace_words(top_uni_simpler, tokenized_stop)
uni_200_tokens = replace_words(uni_200_simpler, tokenized_stop)
top_uni_df['name_tokenized_top_uni'] = top_uni_tokens
uni_200_df['name_tokenized_uni_200'] = uni_200_tokens
```

# 5.  Can you find the best university taking in consideration both rankings? Explain your approach.

We will publish a new ranking aggregated from these two rankings. In order to take in consideration both rankings and find the best university, several methods are available.

1. The simplest one consists in taking the mean from the two original rankings,

2. Another posssibility is to use the harmonic mean. The harmonic mean of 2 values is the same as the mean if the two values are equal. It will be closer to the smaller one if they are different. But in our case we want to give a worst rank for unstable universities. That is why we need to reverse the ranking and compute the "mirrored" harmonic mean.

3. Our final solution is to compute a ranking based on the mean and the use of an $\alpha$ hyperparameter ($0 < \alpha < 1$) so that an "unstable" university (high difference between the two rankings) is more penalized in the final ranking than a stable one.


```python
import matplotlib.pyplot as plt

def mean(a,b):
    return (a + b) / 2

def harmonic(a,b):
    return 2 / (1/a + 1/b)

def mirrored_harmonic(a,b):
    return -1 * harmonic(a,b) + 2 * mean(a,b)

def minus_harmonic(a, b, max_, scale):
    return max_ - harmonic((max_ - a), (max_ - b))

def abs_punish(a, b, alpha):
    return mean(a,b) + alpha * np.abs(a - b)
```

```python
def plot_rank_function(ax, rank_fn, title):
    xs = np.arange(1, 1000)

    min_mean = 2
    max_mean = 16
    step_mean = 2
    means = np.arange(min_mean, max_mean + 1, step_mean)
    ax.set_ylim([0,2 * max_mean])
    ax.set_xlim([0, max_mean + 2 * step_mean])
    ax.set_xticks(means)
    ax.set_yticks(means)
    ax.grid(color='grey', linestyle='dotted', linewidth=1, axis='x')
    ax.set_xlabel('Value of the first rank')
    ax.set_ylabel('Means: arithmetic (straight), {} (curve)'.format(title))

    for i, mean_ in enumerate(means):
        color = 'C{}'.format(i % 10)
        xs2 = 2 * mean_ - xs
        m = mean(xs, xs2)
        inv_h = rank_fn(xs[xs2 > 0], xs2[xs2 > 0])
        ax.plot(xs, m, color);
        ax.plot(xs[xs2 > 0], inv_h, color);


    #ax.set_title('Behavior of the "inverse harmonic mean" for different for different mean');
    ax.set_title('Behavior of the "{}" for different mean'.format(title));
```

```python
# ax1, ax2 = plt.subplot(2, 1, 1)
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,7))
plot_rank_function(ax1, mirrored_harmonic, 'mirrored harmonic')
plot_rank_function(ax2, lambda x, y: abs_punish(x,y, .25), 'abs punish mean')
```

The interpretation of this graph is very meaningful. We see that a university that has rank 8 in both rankings (red line) will be ranked better than a university with a mean rank of 6 (green curve) given by a (10 and a 2). Universities are rewarded for being constant in the different rankings.

This definition of "mirrored harmonic" function is nice, but it has a caveat. Given that two curves can cross, it means that the difference of ranking function is not monotonic. This implies that improving one ranking by 1 place can be punished more than rewarded because the difference between the two rankings increases. This should not happen, that's why we finally picked the "absolute difference punish mean".

Instead we punish the unstable university by an $\alpha=0.25$ which means that having a stable ranking can place a university higher than another with better mean rank but a larger difference. 4 difference rank makes you lose one "mean rank point", e.g. a university having both 8 rank (red horizontal line) will be ranked the same as a university with mean 6 (2 and 10, red diagonal).

```python
new_ranking = rankings_together.copy().loc[:,['rank_uni_200', 'rank_top_uni']]
r_200 = rankings_together.rank_uni_200
r_top = rankings_together.rank_top_uni

def add_ranking(df, on_col):
    rank = df[on_col].sort_values()
    rank = pd.Series(np.arange(1, len(rank) + 1), index=rank.index)
    df[on_col + '_rank'] = rank

new_ranking['mean'] = mean(r_200, r_top)
add_ranking(new_ranking, 'mean')
# new_ranking['harmonic'] = harmonic(r_200, r_top)
# add_ranking(new_ranking, 'harmonic')
new_ranking['abs_punish'] = abs_punish(r_200, r_top, 0.25)
add_ranking(new_ranking, 'abs_punish')
new_ranking['mirrored_harmonic'] = mirrored_harmonic(r_200, r_top)
add_ranking(new_ranking, 'mirrored_harmonic')
```
